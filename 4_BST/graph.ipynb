```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np

file = open("time.txt","r")
Time = file.read()
Time = Time.split(',')
figure(figsize=(12,9), dpi=80)
plt.plot(Time)
x = np.arange(1,100000)
y=[]
for i in x:
    y.append(i*np.log2(i)/15000)
plt.plot(x,y)

plt.legend(["Program Execution Time","nlog(n)"])
plt.ylabel("TIME")
plt.xlabel("N")
plt.title("Lab 4 BST : Insertion time vs nlog(n)")
```
