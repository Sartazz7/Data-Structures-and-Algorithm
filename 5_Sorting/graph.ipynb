```python
import numpy as np
import matplotlib.pyplot as plt
import linecache as lc

N = 1000
swaps = np.zeros((3,3,N))
comp = np.zeros((3,3,N))
time = np.zeros((3,3,N));

def input(line,i,j,a):
    line = line.split(",")
    for k in range(0,N):
        a[i][j][k] = float(line[k])

for i in range(3):
    for j in range(3):
        input(lc.getline("output.txt",9*i+3*j+1),i,j,swaps)
        input(lc.getline("output.txt",9*i+3*j+2),i,j,comp)
        input(lc.getline("output.txt",9*i+3*j+3),i,j,time)
```

```python
for i in range(3):
    for j in range(3):
        print(swaps[i][j])
        print(comp[i][j])
        print(time[i][j])
```

```python
algorithms = ["Quick Sort", "Random Quick Sort", "Bubble Sort", "nlog(n)", "n\u00b2"]
inputs = ["Random input", "Almost sorted (90%)", "Almost sorted reversed"]
x = np.arange(N)+1
```

```python
figure, axis = plt.subplots(1,3, figsize=(16,6))

for i in range(3):
    for j in range(3):
        axis[j].plot(swaps[i][j])

for i in range(3):
    axis[i].plot(x,x*np.log2(x))
    axis[i].plot(x,x*x/4)
    axis[i].set_title(inputs[i])
    axis[i].set_xlabel("N")
    axis[i].legend(algorithms)
    

axis[0].set_ylabel("Number of swaps")
```

```python
figure, axis = plt.subplots(1,3, figsize=(16,6))

for i in range(3):
    for j in range(3):
        axis[j].plot(comp[i][j])

for i in range(3):
    axis[i].plot(x,x*np.log2(x))
    axis[i].plot(x,x*x)
    axis[i].set_title(inputs[i])
    axis[i].set_xlabel("N")
    axis[i].legend(algorithms)
    

axis[0].set_ylabel("Number of comparisions")
```

```python
figure, axis = plt.subplots(1,3, figsize=(16,6))
   
for i in range(3):
    axis[i].plot(time[0][i])
    # Quick Sort
    
for i in range(3):
    axis[i].plot(time[0][i])
    # Random Quick Sort
    
for i in range(3):
    axis[i].plot(time[0][i])
    # Bubble Sort
    

for i in range(3):
    #axis[i].plot(x,x*np.log2(x))
    #axis[i].plot(x,x*x)
    axis[i].set_title(inputs[i])
    axis[i].set_xlabel("N")
    axis[i].legend(algorithms)

axis[0].set_ylabel("Time taken")
```
