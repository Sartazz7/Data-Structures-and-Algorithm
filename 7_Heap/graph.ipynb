```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("heap_data.csv", names=['N', 'Insert Time', 'Insert Swaps', 'Delete Time', 'Delete Swaps'])
```

```python
df.head()
```

```python
plt.figure(figsize=(10,7))
sns.lineplot(x='N', y='Delete Swaps', data=df)
x=np.arange(1000,100000,10)
plt.plot(x,np.log(x)*1.5)
plt.legend(['Average no. of Swaps in delete_min()', 'log(n)'])
```

```python
plt.figure(figsize=(10,7))
sns.lineplot(x='N', y='Insert Swaps', data=df)
x=np.arange(1000,100000,10)
plt.legend(['Average no. of Swaps in insert(int)'])
```

```python
plt.figure(figsize=(10,7))
sns.lineplot(x='N', y='Delete Time', data=df)
x=np.arange(1000,100000,10)
plt.plot(x,np.log(x)*1e-7)
plt.legend(['Average Time Taken in delete_min()', 'log(n)'])
```

```python
plt.figure(figsize=(10,7))
sns.lineplot(x='N', y='Insert Time', data=df)
x=np.arange(1000,100000,10)
plt.plot(x,np.log(x)*1e-8)
plt.legend(['Average Time Taken in insert(int)', 'log(n)'])
```
